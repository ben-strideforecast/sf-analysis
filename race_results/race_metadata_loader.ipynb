# Notebook B — Race Metadata Loader
Load the race discovery CSV from Notebook A, filter it, and prepare a clean table for metadata loading.

```python
import pandas as pd
from pathlib import Path
```

## Configuration

```python
# Path to the race_discovery CSV from Notebook A
# Leave as None to auto-detect the most recent file, or set explicitly:
#   DISCOVERY_CSV = Path('outputs/race_discovery/race_discovery_20260509.csv')
DISCOVERY_CSV = Path('outputs/race_discovery/race_discovery_20260509.csv')
```

## Load Discovery CSV

```python
discovery_dir = Path('outputs/notebook_a_race_discovery')

if DISCOVERY_CSV:
    csv_path = Path(DISCOVERY_CSV)
else:
    candidates = sorted(discovery_dir.glob('race_discovery_*.csv'), reverse=True)
    if not candidates:
        raise FileNotFoundError(f'No race_discovery_*.csv found in {discovery_dir}')
    csv_path = candidates[0]

print(f'Loaded: {csv_path}')
df = pd.read_csv(csv_path)
print(f'Rows: {len(df)}')
df
```

## Filter

```python
df_filtered = df.copy()

# Add filters here, e.g.:
# df_filtered = df_filtered[df_filtered['distance'] == '5K']

print(f'Rows after filtering: {len(df_filtered)}')
df_filtered
```

## Prepare for Metadata Load

```python
keep_cols = [c for c in ['race_id', 'race_name', 'event_id', 'event_name', 'distance',
                          'start_date', 'city', 'state', 'zipcode', 'race_url']
             if c in df_filtered.columns]

df_metadata = df_filtered[keep_cols].copy()
df_metadata['race_id'] = pd.to_numeric(df_metadata['race_id'], errors='coerce').astype('Int64')

if 'event_id' in df_metadata.columns:
    df_metadata['event_id'] = pd.to_numeric(df_metadata['event_id'], errors='coerce').astype('Int64')

print(f'Ready to load: {len(df_metadata)} rows, {len(df_metadata["race_id"].unique())} races')
df_metadata
```
