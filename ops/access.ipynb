```python
import pyodbc
import pandas as pd

conn_str = (
    r'DRIVER={Microsoft Access Driver (*.mdb, *.accdb)};'
    r'DBQ=C:\Users\Ben Sylve\Documents\git_sf\sf-pipeline\db\sf_race_metadata.accdb;'
)

conn = pyodbc.connect(conn_str)
cursor = conn.cursor()

tables = [row.table_name for row in cursor.tables(tableType='TABLE')]
print(tables)

# Loop through each table and collect column metadata
schema = []
for table in tables:
    columns = cursor.columns(table=table)
    for col in columns:
        schema.append({
            'table_name': col.table_name,
            'column_name': col.column_name,
            'data_type': col.type_name,
            'max_length': col.column_size,
            'nullable': col.nullable,
            'ordinal_position': col.ordinal_position
        })

df = pd.DataFrame(schema)
print(df)
```
