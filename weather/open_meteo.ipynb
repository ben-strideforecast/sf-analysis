```python
from datetime import datetime
import meteostat as ms
from geopy.geocoders import Nominatim

geolocator = Nominatim(user_agent="stride_forecast")
address = "Jackson, MS"

location = geolocator.geocode(address)

lat = location.latitude
lon = location.longitude

POINT = ms.Point(lat, lon)

start = datetime(2025, 7, 4, 7, 0)
end   = datetime(2025, 7, 4, 15, 0)

stations = ms.stations.nearby(POINT, limit=4)
ts = ms.hourly(stations, start, end)
df = ms.interpolate(ts, POINT).fetch()

df['temp_f'] = df['temp'] * 9/5 + 32

# wdir = wind direction in degrees, measured clockwise from north
# representing where the wind is COMING FROM:
#   0/360 = North, 90 = East, 180 = South, 270 = West
# 2024 GCMR showed 40-60 degrees all morning = NE wind (cold, off-shore)

print(df.to_string())
```
