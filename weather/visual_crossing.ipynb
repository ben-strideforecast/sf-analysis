```python
"""
Visual Crossing API - Hourly weather for a zip code on a given date.
Free tier: 1000 requests/day
Free key: https://www.visualcrossing.com/sign-up
"""

import requests
import pandas as pd

API_KEY  = "PK83JKSM22UETTX5DK9XHTNK5"
ZIP_CODE = "39216"          # Gulfport, MS
DATE     = "2025-07-04"

# ── Pull ──────────────────────────────────────────────────────────────────────

url = f"https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/timeline/{ZIP_CODE}/{DATE}"

resp = requests.get(url, params={
    "key":          API_KEY,
    "unitGroup":    "us",           # Fahrenheit, mph, inches
    "include":      "hours",
    "contentType":  "json",
})
resp.raise_for_status()

data = resp.json()

# ── Parse ─────────────────────────────────────────────────────────────────────

hours = data["days"][0]["hours"]

df = pd.DataFrame(hours)

print(f"Location: {data['resolvedAddress']}")
print(df.to_string(index=False))
```
